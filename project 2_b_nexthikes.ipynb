```python
plt.figure(figsize=(10,5))
hourly_rentals=merged_df2.groupby("hr")[["cnt","casual","registered"]].mean()
plt.plot(hourly_rentals.index,hourly_rentals["cnt"],linestyle="--",color="blue",label="Total rentals")
plt.plot(hourly_rentals.index,hourly_rentals["casual"],linestyle="--",color="red",label="Casual users")
plt.plot(hourly_rentals.index,hourly_rentals["registered"],linestyle="--",color="green",label="Registered Users")
plt.xlabel("Hour of the day")
plt.ylabel("Average bike rentals")
plt.title("Hourly trend of bike rentals")
plt.xticks(range(0,24))
plt.legend()
plt.grid(True)
plt.show()
```

## There are two major spikes in bike rentals:
##  Morning peak(around 8 AM):due to commuters travelling to school or office
##    Evening Peak (around5-7 PM):due to people returning home from office
##    Rentals reman low during midday and late night
##    registered users are increasing in morning and evening
##    Casual riders are higher in the afternoon (10 AM -6PM),suggesting more leisure riders.
##    Rentals drop significantly after 9 PM, likely due to fewer people riding bikes late at night
